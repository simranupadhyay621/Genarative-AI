```python
#26.Basic Phrase Matching "Machine learning is part of artificial intelligence" "machine learning", "artificial intelligence"

import spacy
from spacy.matcher import PhraseMatcher

nlp=spacy.load('en_core_web_sm')
matcher=PhraseMatcher(nlp.vocab)

phrase=["machine learning", "artificial intelligence"]
pattern=[nlp(text) for text in phrase]

matcher.add("TechTerm",pattern)

doc=nlp(u"Machine learning is part of artificial intelligence")
matches=matcher(doc)

for match_id, start,end in matches:
  span=doc[start:end]
  print("Matched Phrase:", span.text)
```

```python
#27.Match Product Names "I bought an iPhone yesterday" "iPhone", "Samsung Galaxy"

import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")

doc = nlp("I bought an iPhone yesterday")

matcher = PhraseMatcher(nlp.vocab)

products = ["iPhone", "Samsung Galaxy"]

patterns = [nlp(product) for product in products]

matcher.add("PRODUCTS", patterns)

matches = matcher(doc)

for match_id, start, end in matches:
    span = doc[start:end]
    print(span.text)
```

```python
#28.Match City Names "Delhi is a busy city" "Delhi", "Mumbai"

doc=nlp("Delhi is a busy city")
matcher=PhraseMatcher(nlp.vocab)

city=["Delhi", "Mumbai"]
pattern=[nlp(city) for city in city]
matcher.add("CITY",pattern)

matches=matcher(doc)

for match_id,start,end in matches:
  span=doc[start:end]
  print(span.text)
```

```python
#29.Count Phrase Matches "Python is used in deep learning" "deep learning", "python"

matcher = PhraseMatcher(nlp.vocab, attr="LOWER")

phrases = ["deep learning", "python"]
patterns = [nlp.make_doc(text) for text in phrases]

matcher.add("TechTerms", patterns)

doc = nlp("Python is used in deep learning")
matches = matcher(doc)


print("Total Matches:", len(matches))

for match_id, start, end in matches:
    span = doc[start:end]
    print("Matched Phrase:", span.text)

```
