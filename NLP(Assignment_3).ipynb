```python
#1. Create a Doc object from the file peterrabbit.txt

import spacy
nlp=spacy.load('en_core_web_sm')

with open("peterrabbit.txt", "r", encoding="utf-8") as f:
  text=f.read()

doc=nlp(text)
print("Number of tokens:", len(doc))
print("First 10 tokens:", [token.text for token in doc[:10]] )
```

```python
#2. For every token in the third sentence, print the token text, the POS tag, the fine-grained TAG tag, and the description of the fine-grained tag.

doc=nlp(text)

sentences=list(doc.sents)

third_sentence=sentences[2]
for token in third_sentence:
  print(f"Text: {token.text},"
        f"POS: {token.pos_},"
        f"TAG: {token.tag_},"
        f"Discription: {spacy.explain(token.tag_)}")
```

```python
#3. Provide a frequency list of POS tags from the entire document

import spacy
from collections import Counter

nlp=spacy.load('en_core_web_sm')

with open("peterrabbit.txt", "r", encoding="utf-8") as f:
  text=f.read()
doc=nlp(text)

pos_tag=[token.pos_ for token in doc]

pos_freq=Counter(pos_tag)
print("Frequency of POS tags in the document:")
for pos, freq in pos_freq.items():
    print(f"{pos}: {freq}")
```

```python
#4. CHALLENGE: What percentage of tokens are nouns?

doc=nlp(u"Anwesha is a nlp teacher.")

noun_count=sum(1 for token in doc if token.pos_=="NOUN" )
print("Noun_count:",noun_count)

total_token=len(doc)
print("Total_token:",total_token)

percentage=(noun_count/total_token)*100
print("Percentage:",percentage)

```

```python
#5.Display the Dependency Parse for the third sentence.

import spacy

nlp = spacy.load("en_core_web_sm")

text = "Python is easy. NLP is interesting. AI changes the world."
doc = nlp(text)

third_sentence = list(doc.sents)[2]

for token in third_sentence:
    print(token.text, "->", token.dep_, "->", token.head.text)

```

```python
#6.Show the first two named entities from Beatrix Potter's The Tale of Peter Rabbit

doc=nlp(u"Beatrix Potter's The Tale of Peter Rabbit")

for ent in doc.ents[:2]:
  print(ent.text, "-", ent.label_, "-", spacy.explain(ent.label_))
```

```python
#7. How many sentences are contained in The Tale of Peter Rabbit?

doc=nlp(u"The Tale of Peter Rabbit?")

sentences=[sent.text for sent in doc.sents]
print("Sentence count:",len(sentences))
print(sentences)
```

```python
#8.CHALLENGE: How many sentences contain named entities?

import spacy

nlp = spacy.load("en_core_web_sm")

text = "Virat Kohli plays for India. He is a famous cricketer. Cricket is popular worldwide."

doc = nlp(text)

sentences_with_entities = [sent for sent in doc.sents if any(ent in sent.ents for ent in doc.ents)]
count = len(sentences_with_entities)

print("Sentences containing named entities:", count)
for sent in sentences_with_entities:
    print("-", sent.text)
```
